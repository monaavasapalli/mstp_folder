```python
x,y
```

```python
x_mean=x.mean()
y_mean=y.mean()
print(x_mean,y_mean)
```

```python
num=0
den=0
for i in range(0,len(x)):
    num+=(x[i]-x_mean)*(y[i]-y_mean)
    den+=(x[i]-x_mean)**2
c=num/den
print(c)
C=y_mean-(c*x_mean)
print(C)
```

```python
plt.scatter(c,C,color='b')
plt.grid()
plt.show()
```

```python
y_pred=c*x+c
print(y_pred)
plt.scatter(brain['Head Size(cm^3)'],brain['Brain Weight(grams)'],color='c')
y_pred=c*x+c
```

```python
plt.scatter(brain['Head Size(cm^3)'],brain['Brain Weight(grams)'],brain['Age Range'])
plt.plot(brain['Head Size(cm^3)'],y_pred,'c',label='Regression line')
plt.xlabel('Head Size(cm^3)')
plt.ylabel('Brain Weight(grams)')
plt.legend()
plt.show()
```
