```python
import pandas as pd
shirt = pd.DataFrame({'Height':H,'Weight':W,'size':T})
shirt
```

```python
((shirt['Height'] - H1)**2 +(shirt['Weight']-W1)**2)**0.5
```

```python
import seaborn as sns

sns.scatterplot(x = shirt['Height'],y = shirt['Weight'],hue = shirt['size'])
sns.scatterplot(x = [161],y = [61])
```

```python
import matplotlib.pyplot as plt

plt.scatter(shirt['Height'],shirt['Weight'])
plt.scatter(161,61,color='r',marker='.')
```

```python
X = shirt[['Height','Weight']]
Y = shirt['size']
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.3,random_state = 42)

knn = KNeighborsClassifier(n_neighbors = 3)

knn.fit(X_train,Y_train)
```

```python
knn.predict([[161,61]])
```

```python
from sklearn.metrics import accuracy_score

Y_pred = knn.predict(X_test)
accuracy_score(Y_test,Y_pred)
```
